```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="/content/dataset/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    name="robot_detector"
)
```

```python
from ultralytics import YOLO

model = YOLO("/content/runs/detect/robot_detector/weights/best.pt")

results = model.predict(
    source="/content/video.mp4",
    save=True,
    conf=0.5
)
```
